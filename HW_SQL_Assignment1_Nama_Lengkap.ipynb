Pertanyaan 3

Create table bernama dim_customer_(last name kamu) dan isi datanya dengan data pada dim customer yang memiliki gender female

```python
%%bigquery --project avid-compound-391908

create table dibimbing.dim_customer_agung as
select *
from dibimbing.dim_customer
where gender = 'F'
```
